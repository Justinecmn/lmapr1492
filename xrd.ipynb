```python
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester ("v1d7ZHuxlIRyoKmi") as m:
    
    #Création de la structure du cristal à partir du MP
    structure = m.get_structure_by_material_id("mp-961685")
```

```python
xrd = XRDCalculator(wavelength='CuKa')
xrd_pattern = xrd.get_pattern(structure)
```

```python
pic_1 = "Le pic de diffraction 1 est associé à : \n" 
pic_2 = "Le pic de diffraction 2 est associé à : \n"
pic_3 = "Le pic de diffraction 3 est associé à : \n"
print(pic_1,
      "\nAngle 2Θ = ", xrd_pattern.x[1], 
      "°\nIntensity = ", xrd_pattern.y[1], 
      "%\nIndices h k l : ", xrd_pattern.hkls[1], "\n")
print(pic_2,
      "\nAngle 2Θ = ", xrd_pattern.x[2], 
      "°\nIntensity = ", xrd_pattern.y[2], 
      "%\nIndices h k l : ", xrd_pattern.hkls[2],"\n")
print(pic_3,
      "\nAngle 2Θ = ", xrd_pattern.x[3], 
      "°\nIntensity = ", xrd_pattern.y[3], 
      "%\nIndices h k l : ", xrd_pattern.hkls[3],"\n")
    
```
