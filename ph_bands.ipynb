```python
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.phonon import plotter
from pymatgen.phonon import bandstructure
from IPython.display import Image
import matplotlib.pyplot as plt
import numpy as np
from math import pi
from math import fabs

with MPRester ("v1d7ZHuxlIRyoKmi") as m:
    
    #Création de la structure du cristal et de la structure de bande de phonons à partir du MP
    structure   = m.get_structure_by_material_id("mp-961685")
    phonons   = m.get_phonon_bandstructure_by_material_id("mp-867339") 
```

```python
phonon_plot = plotter.PhononBSPlotter(phonons)
phonon_plot.plot_brillouin() 
```

```python
#la structure de bande des phonons

phonon_plot.get_plot()
plt.title('La structure de la bande des phonons')
```

```python
#récupération des données
bs_plot_data_phonon = phonon_plot.bs_plot_data()
```

```python
#pour chaque chemin de la zone de Brillouin
phonon_plot.get_plot()
plt.title('Réprésentation des 3 premières branches acoustiques qui vont être analysées')
for i in range(len(bs_plot_data_phonon.get('distances'))):
    plt.plot(bs_plot_data_phonon.get('distances')[i],bs_plot_data_phonon.get('frequency')[i][0],'green')  #On trace la première branche acoustique
    plt.plot(bs_plot_data_phonon.get('distances')[i],bs_plot_data_phonon.get('frequency')[i][1],'red')    #On trace la deuxième branche acoustique
    plt.plot(bs_plot_data_phonon.get('distances')[i],bs_plot_data_phonon.get('frequency')[i][2],'orange') #On trace la troisième branche acoustique
```

```python
#Γ-X
vs_gamma_x         = (((bs_plot_data_phonon.get('frequency')[0][0][2] - bs_plot_data_phonon.get('frequency')[0][0][0]))*(10**12)*(2*np.pi))/(bs_plot_data_phonon.get('distances')[0][2] - bs_plot_data_phonon.get('distances')[0][0])
vitesseSon_gamma_x = vs_gamma_x/(10**(10)) #car les vecteurs d'onde sont exprimés en Å 
print("La vitesse du son d'une branche acoustique pour le chemin Γ-X vaut : ", vitesseSon_gamma_x, "m/s.")

#Γ-K
vs_gamma_k         = (((bs_plot_data_phonon.get('frequency')[3][0][-3]- bs_plot_data_phonon.get('frequency')[3][0][-1]))*(10**12)*(2*np.pi))/(bs_plot_data_phonon.get('distances')[3][-3]- bs_plot_data_phonon.get('distances')[3][-1])
vitesseSon_gamma_k = vs_gamma_k/(10**(10)) #car les vecteurs d'onde sont exprimés en Å 
print("La vitesse du son d'une branche acoustique pour le chemin Γ-K vaut : ", vitesseSon_gamma_k, "m/s.")

#Γ-L
vs_gamma_l         = (((bs_plot_data_phonon.get('frequency')[4][0][2] - bs_plot_data_phonon.get('frequency')[4][0][0]))*(10**12)*(2*np.pi))/(bs_plot_data_phonon.get('distances')[4][2] - bs_plot_data_phonon.get('distances')[4][0]) 
vitesseSon_gamma_l = vs_gamma_l/(10**(10)) #car les vecteurs d'onde sont exprimés en Å 
print("La vitesse du son d'une branche acoustique pour le chemin Γ-L vaut : ", vitesseSon_gamma_l, "m/s.")

```
